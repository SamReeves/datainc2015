```python
import numpy as np
import math
import pandas as pd
import json
```

```python
#Q1   A: 0.43423286846573694

#What is the probability that the tourist is at least 
#3 city blocks (as the crow flies) from Broadway and Broadway after 10 moves?

F_D10 = pd.DataFrame(pd.read_json('F_D10.json'))
```

```python
def threeBlocks():
    n = 0
    t = 0
    
    for x in xrange(499999):
        
        if F_D10.iloc[t][0] >= 3.0:
            n += 1
            
        t += 1
        
    return float(n/500000.0)

threeBlocks()
```

```python
#Q2    A: 0.0294064
#What is the probability that the tourist is ever at least 
#5 city blocks (as the crow flies) from Broadway and Broadway within 10 moves?

bank10 = pd.DataFrame(pd.read_json('bank10.json'))
```

```python
def locationList():
    bigBook = []
    trial = 0
    
    for _ in xrange(500000):
        step = 0
        smallBook = []
        
        for item in xrange(10):
            move = bank10.iloc[trial,step]
            
            if len(smallBook) == 0:
                location = move
            else:
                oldLoc = smallBook[-1]
                newx = oldLoc[0] + move[0]
                newy = oldLoc[1] + move[1]
                location = [newx, newy]
                
            smallBook.append(location)
            step += 1
            
        bigBook.append(smallBook)
        trial += 1
        
    return bigBook
```

```python
with open('allLoc10.json', 'w') as outfile:
    json.dump(locationList(), outfile)
```

```python
allLoc10 = pd.DataFrame(pd.read_json('allLoc10.json'))
allLoc10
```

```python
def distanceList():
    bigBook = []
    trial = 0
    
    for _ in xrange(500000):
        smallBook = []
        step = 0
        
        for item in xrange(10):
            loc = allLoc10.iloc[trial][step]
            x = loc[0]**2
            y = loc[1]**2
            
            dist = math.sqrt(x + y)
            step +=1
            smallBook.append(dist)
            
        trial += 1
        bigBook.append(smallBook)
               
    return bigBook
```

```python
A_D10 = pd.DataFrame(distanceList())
```

```python
def fiveBlocks():
    n = 0.0
    trial = 0
    
    for row in xrange(500000):
        step = 0
        for item in xrange(10):
            dist = A_D10.iloc[trial][step]
            if dist >= 5.0:
                n += 1
            else:
                n += 0
            step += 1
        trial += 1
    
    return n
```

```python
float(fiveBlocks()/(5000000.0))
```
